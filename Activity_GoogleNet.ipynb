# 모델 구현

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, Model
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.utils import to_categorical
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
```

```python
def inception_block(x, filters_1x1, filters_3x3_reduce, filters_3x3,
                   filters_5x5_reduce, filters_5x5, filters_pool_proj, name=None):
    """
    Inception 블록 구현
    - 1x1, 3x3, 5x5 conv와 max pooling을 병렬로 수행
    - 서로 다른 스케일의 특징을 동시에 추출
    """
    # 1x1 convolution branch
    conv_1x1 = layers.Conv2D(filters_1x1, (1, 1), padding='same', activation='relu',
                            name=f'{name}_1x1')(x)

    # 3x3 convolution branch (1x1로 차원 축소 후)
    conv_3x3_reduce = layers.Conv2D(filters_3x3_reduce, (1, 1), padding='same', activation='relu',
                                   name=f'{name}_3x3_reduce')(x)
    conv_3x3 = layers.Conv2D(filters_3x3, (3, 3), padding='same', activation='relu',
                            name=f'{name}_3x3')(conv_3x3_reduce)

    # 5x5 convolution branch (1x1로 차원 축소 후)
    conv_5x5_reduce = layers.Conv2D(filters_5x5_reduce, (1, 1), padding='same', activation='relu',
                                   name=f'{name}_5x5_reduce')(x)
    conv_5x5 = layers.Conv2D(filters_5x5, (5, 5), padding='same', activation='relu',
                            name=f'{name}_5x5')(conv_5x5_reduce)

    # MaxPooling branch
    pool = layers.MaxPool2D((3, 3), strides=(1, 1), padding='same', name=f'{name}_pool')(x)
    pool_proj = layers.Conv2D(filters_pool_proj, (1, 1), padding='same', activation='relu',
                             name=f'{name}_pool_proj')(pool)

    # 모든 branch를 concatenate
    output = layers.Concatenate(axis=3, name=f'{name}_concat')([conv_1x1, conv_3x3, conv_5x5, pool_proj])

    return output
```

```python
def auxiliary_classifier(x, num_classes, name):
    """
    보조 분류기 구현
    - 네트워크 중간에서 분류를 수행
    - Gradient vanishing 문제 해결을 위함
    """
    # Average pooling으로 feature map 크기 감소
    x = layers.AveragePooling2D((5, 5), strides=(3, 3), name=f'{name}_pool')(x)

    # 1x1 conv로 채널 수 조정
    x = layers.Conv2D(128, (1, 1), padding='same', activation='relu', name=f'{name}_conv')(x)

    # Fully connected layers
    x = layers.Flatten(name=f'{name}_flatten')(x)
    x = layers.Dense(1024, activation='relu', name=f'{name}_fc1')(x)
    x = layers.Dropout(0.7, name=f'{name}_dropout')(x)  # 높은 dropout rate

    # 최종 분류층
    x = layers.Dense(num_classes, activation='softmax', name=f'{name}_classifier')(x)

    return x
```

```python
def GoogleNet(input_shape=(32, 32, 3), num_classes=10):
    """
    GoogleNet (Inception v1) 모델 구현
    - 보조 분류기 2개 포함
    - CIFAR-10에 맞게 크기 조정
    """
    input_tensor = layers.Input(shape=input_shape)

    # ========== Initial Convolution Layers ==========
    # 원본 GoogleNet은 224x224 입력이지만, CIFAR-10은 32x32이므로 조정
    x = layers.Conv2D(64, (3, 3), strides=(1, 1), padding='same', activation='relu',
                     name='conv1_3x3_s1')(input_tensor)
    x = layers.BatchNormalization(name='conv1_bn')(x)
    x = layers.MaxPool2D((2, 2), strides=(2, 2), padding='same', name='pool1_2x2_s2')(x)

    # 추가 conv layers
    x = layers.Conv2D(64, (1, 1), strides=(1, 1), padding='same', activation='relu',
                     name='conv2_1x1_reduce')(x)
    x = layers.Conv2D(192, (3, 3), strides=(1, 1), padding='same', activation='relu',
                     name='conv2_3x3')(x)
    x = layers.BatchNormalization(name='conv2_bn')(x)
    x = layers.MaxPool2D((2, 2), strides=(2, 2), padding='same', name='pool2_2x2_s2')(x)

    # ========== Inception Blocks 3a, 3b ==========
    x = inception_block(x, 64, 96, 128, 16, 32, 32, name='inception_3a')
    x = inception_block(x, 128, 128, 192, 32, 96, 64, name='inception_3b')
    x = layers.MaxPool2D((2, 2), strides=(2, 2), padding='same', name='pool3_2x2_s2')(x)

    # ========== Inception Blocks 4a~4e ==========
    x = inception_block(x, 192, 96, 208, 16, 48, 64, name='inception_4a')

    #  첫 번째 보조 분류기 (4a 이후)
    aux1 = auxiliary_classifier(x, num_classes, 'aux1')

    x = inception_block(x, 160, 112, 224, 24, 64, 64, name='inception_4b')
    x = inception_block(x, 128, 128, 256, 24, 64, 64, name='inception_4c')
    x = inception_block(x, 112, 144, 288, 32, 64, 64, name='inception_4d')

    #  두 번째 보조 분류기 (4d 이후)
    aux2 = auxiliary_classifier(x, num_classes, 'aux2')

    x = inception_block(x, 256, 160, 320, 32, 128, 128, name='inception_4e')

    # ========== Inception Blocks 5a, 5b ==========
    x = inception_block(x, 256, 160, 320, 32, 128, 128, name='inception_5a')
    x = inception_block(x, 384, 192, 384, 48, 128, 128, name='inception_5b')

    # ========== Final Classification Layer ==========
    x = layers.GlobalAveragePooling2D(name='pool5_gap')(x)
    x = layers.Dropout(0.4, name='dropout')(x)
    main_output = layers.Dense(num_classes, activation='softmax', name='main_classifier')(x)

    # 🎯 3개의 출력을 가진 모델 생성
    # [메인 분류기, 보조 분류기 1, 보조 분류기 2]
    model = Model(inputs=input_tensor, outputs=[main_output, aux1, aux2], name='GoogleNet')

    return model
```
