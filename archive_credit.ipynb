```python
from sklearn.preprocessing import StandardScaler

data = df.drop(['Time'], axis=1)

data['Amount'] = StandardScaler().fit_transform(data['Amount'].values.reshape(-1, 1))
```

```python
RANDOM_SEED = 42
```

```python
X_train, X_test = train_test_split(data, test_size=0.2, random_state=RANDOM_SEED)
X_train = X_train[X_train.Class == 0]
X_train = X_train.drop(['Class'], axis=1)

y_test = X_test['Class']
X_test = X_test.drop(['Class'], axis=1)

X_train = X_train.values
X_test = X_test.values

X_train.shape
```

## 오토인코더

입력층 (31차원)

    ↓

Dense(20) - 31개 정보를 20개로 축소

    ↓

Dense(12) - 20개 정보를 12개로 축소  

    ↓

Dense(8)  - 12개 정보를 8개로 최종 압축 ← 이게 encoding_dim!

    ↓

[복원 과정]

    ↓

Dense(12) - 8개에서 12개로 복원

    ↓  

Dense(20) - 12개에서 20개로 복원

    ↓
    
출력층 (31차원) - 20개에서 31개로 완전 복원

```python
input_dim = X_train.shape[1]
encoding_dim = 8 # 인코딩 차원을 설정 (임의로 설정한 값, 조정 가능)
input_layer = Input(shape=(input_dim,))

# # 인코더
# encoded = Dense(encoding_dim, activation='relu')(input_layer)
# encoded = Dense(int(encoding_dim / 2), activation='relu')(encoded)

# # 디코더
# decoded = Dense(int(encoding_dim / 2), activation='relu')(encoded)
# decoded = Dense(input_dim, activation='sigmoid')(decoded)

# 인코더: 점진적 축소 (31 → 20 → 12 → 8)
encoded = Dense(20, activation='relu')(input_layer)      # 31 → 20
encoded = Dense(12, activation='relu')(encoded)          # 20 → 12
encoded = Dense(encoding_dim, activation='relu')(encoded) # 12 → 8 (병목층)

# 디코더: 점진적 복원 (8 → 12 → 20 → 31)
decoded = Dense(12, activation='relu')(encoded)          # 8 → 12
decoded = Dense(20, activation='relu')(decoded)          # 12 → 20
decoded = Dense(input_dim, activation='sigmoid')(decoded) # 20 → 31


autoencoder = Model(inputs=input_layer, outputs=decoded)
autoencoder.compile(optimizer=Adam(learning_rate=0.001), loss='mse')
```

```python

# 3. 모델 학습
# 훈련데이터의 0.1 을 validation data로 사용
history = autoencoder.fit(X_train, X_train,
                          epochs=50,
                          batch_size=256,
                          shuffle=True,
                          validation_split = 0.1)
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper right');
```

```python

# X_test 값을 넣어서 Reconstruction Error 값을 얻고,
# 실제 레이블인 y_test와 데이터프레임으로 구성
# Reconstruction Error + 실제 레이블의 형태가 됨

predictions = autoencoder.predict(X_test)
mse = np.mean(np.power(X_test - predictions, 2), axis=1)
error_df = pd.DataFrame({'reconstruction_error': mse,
                        'true_class': y_test})
error_df.describe()
```

```python
threshold = 15
```

```python

groups = error_df.groupby('true_class')
fig, ax = plt.subplots()

for name, group in groups:
    ax.plot(group.index, group.reconstruction_error, marker='o', ms=3.5, linestyle='',
            label= "Fraud" if name == 1 else "Normal")
ax.hlines(threshold, ax.get_xlim()[0], ax.get_xlim()[1], colors="r", zorder=100, label='Threshold')
ax.legend()
plt.title("Reconstruction error for different classes")
plt.ylabel("Reconstruction error")
plt.xlabel("Data point index")
plt.show();
```

```python
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns

y_pred = [1 if e > threshold else 0 for e in error_df.reconstruction_error.values]
conf_matrix = confusion_matrix(error_df.true_class, y_pred)
plt.figure(figsize=(8, 8))
sns.heatmap(conf_matrix, xticklabels=LABELS, yticklabels=LABELS, annot=True, fmt="d");
plt.title("Confusion matrix")
plt.ylabel('True class')
plt.xlabel('Predicted class')
plt.show()

print(classification_report(error_df.true_class, y_pred))
```
